# DataFrame structure intro




```
import pandas as pd
record1=pd.Series({'Name': 'Alice',
                   'Class': 'Physics',
                   'Score': 85})
record2=pd.Series({'Name':'Jack',
                   'Class':'Chemistry',
                   'Score':82})
record3=pd.Series({'Name': 'Helen',
                   'Class':'Biology',
                   'Score':'90'}) 
```

```
##Asi como Series, DataFrame tambien es un objeto indexado. Por lo que vamos a usar un grupo de series
#donde cada serie representa una fila, como si se tratara de una serie, tambien podemos pasarle
#nuestro datos individuales en un array, y los valores indexados en un segundo argumento

df=pd.DataFrame([record1,record2,record3],index=['School1','School2','School1'])
df.head()
```

```
##Asi mismo de manera alternativa podemos crear el DataFrame con una lista de diccionarios
students=[{'Name': 'Alice',
                   'Class': 'Physics',
                   'Score': 85},
          {'Name':'Jack',
           'Class':'Chemistry',
           'Score':82},
          {'Name': 'Helen',
           'Class':'Biology',
           'Score':90}]
df=pd.DataFrame(students,index=['School1','School2','School1'])           
df.head()
```

```
##Podemos usar el atributo Loc, para conseguir los datos de una fila, como haremos con School2

df.loc['School2']
```

```
##Como podemos notar el nombre de la serie es returnada como el valor de indexacion
#Podemos verificar el tipo de dato con type function ejemplo:
type(df.loc['School2'])
```

```
##Cuando el valor de indexacion esta repetido, no nos devuelve una serie si no una dataframe
#Ejemplo
df.loc['School1']
```

```
type(df.loc['School1'])
```

```
##Una de las ventajas es que al utilizar pandas, podemos rapidamente seleccionar bases de datos
#con multiple ejes, por ejemplo, si solo queremos el nombre de los estudiantes de la escuela 1
#solo necesitamos 2 parametros the row and the column name
df.loc['School1','Name']
```

```
##Recordemos que los desarrolladores de pandas han implementado esto utilizando operadores de 
#indexacion(Indexing) y no parametros de funcion 
#Entonces como seleccionariamos una simple columna? hay pocos mecanismos para desarrollarlo
#una es transponiendo la matriz con T attribute
df.T
```

```
df.T.loc['Name']
```

```
##Aunque iloc y loc se usa para indexacion de filas, Pandas se reserva el uso del operador de indexacion
#para la seleccion de columnas. En pandas, las columnas siempre tienen un nombre
df['Name']
```

```
##En la practica esto funciona muy bien para agregar o eliminar nuevas columnas.
#Sin embargo, esto tambien indica que que si que si trata de usar una key, produce un error
#Ejemplo
df.loc['Name']
```

```
#Es de notar tambien que el resultado de una sola columna, es un objeto tipo serie
type(df['Name'])
```

```
#Dado que el operador de indexacion nos puede dar como resultado dataFrame o series
#se pueden encadenar operaciones juntas. Por ejemplo, proyectar las filas con School1
#y luego la columna con solo los nombres
df.loc['School1']['Name']
```

```
##El encadenamiento por indexacion en el retorno de otro indice, puede venir con algun costo
#y se evita mejor si trata de utilizar otro enfoque(approach). En particular, el encadenamiento
#tiende a generar que pandas una copia del dataFrame en lugar de una vista del DataFrame.
#Para seleccionar datos, esto no es un gran problema, aunque puede ser un poco más lento de lo 
#necesario. Sin embargo, si se estan modificando datos, es una distinción importante, puede generar
#errores
```

```
##Con otro enfoque. loc tambien permite el corte(slicing)
#Si quisieramos todas las filas, podemos utilizar colon operator(:) para obtener full slice
#desde el principio al final, como en una lista de python. Y entonces, podemos agregar una
#el nombre a la columna, como segundo parametro como un string. Si queremos incluir multiples
#columnas, podriamos hacerlo en una lista. y PANDAS nos devolvería todas las columnas pedidas.
df.loc[:,['Name','Score']]
```

```
##La clave en los conceptos, es que tomar las columnas y filas son solo para nuestro beneficio, 
#debajo de todo hay una matriz etiquetada de dos ejes, transponerla es facil. Tambien considerar
#el tema de encadenamiento cuidadosamente y tratar de evitarlo, puede causar algunos resultados
#impredecibles, donde se intentaba obtener una vista de los datos y Pandas entregó una copia del dataframe
```

```
##Ahora como eliminar datos de un dataFrame?
#Mediante la funcion drop. Esta función toma un solo parametro, ya sea el index o la etiqueta de la fila
#Nuevamente la función drop no nos devuelve el DataFrame original con la fila eliminada, en realidad nos 
#devuelve una copia del DataFrame con la fila eliminada.
df.drop('School1')
```

```
#pero veamos el dataFrame original
df
```

```
##Drop tiene dos parametros opcionales muy interesantes, el primero es:
#inplace, en caso de configurarlo en True, el DataFrame se actualizara,
#en vez de retornar una copia 
#El segundo parametro es axes. Que por defecto tiene el valor de 0, indicando que el eje es una fila,
#pero se puede cambiar a 1 si se quiere eliminar una columna, por ejemplo:
copy_df=df.copy()
#vamos a eliminar la columna Name en esta copia 
copy_df.drop('Name',inplace=True,axis=1)
copy_df
```

```
##Hay otra manera de eliminar una columna, y es mediante el uso directo del indexing operator
#usando una keyword. This way dropping data, sin embargo, tiene efecto inmediato y no hay 
#retorno a la vista, Ejemplo
del copy_df['Class']
copy_df
```

```
##Finalmente, para agregar una columna a un DataFrame, es simplemente 
#agregar valores a un operador de indexacion, Ejemplo
df['ClassRanking']=None
df
```

# Querying DataFrame (consultando en DataFrame)

```
#Lo primero para entender sobre consultas en DataFrame es entender el enmarascamiento Booleano
#El enmascaramiento de Booleano es el corazon de las consultas, en numpy y pandas
##Un enmascaramiento booleano es un arreglo de una o dos dimensiones, este array(arreglo) se superpone
#donde en el lugar que tome valores verdaderos retornara ese valor #Ejemplo, comencemos importando un
#dataset de admitidos a graduacion
import pandas as pd
df=pd.read_csv('/content/drive/MyDrive/Python_for_Data_Science/Week_2/datasets/Admission_Predict.csv',index_col=0)
#Ahora vamos a limpiar las bases y los nombres mal puestos
df.columns=[x.lower().strip() for x in df.columns]
df.head()
```

```
## Las mascaras booleanas se crean aplicando operadores directamente sobre las series de Panda o DataFrames
#Por ejemplo, vamos a ver solo los estudiantesque tienen una probabilidad mayor a 0.7
#Vamos proyectar la mascara booleana sobre la columna que necesitamos usando el operador de indexacion
#con el operador de que sea mayor a 0.7, asi que a continuacion crearemos la mascara de admision.
admit_mask=df['chance of admit']>0.7
admit_mask
```

```
##Por debajo Pandas esta aplicando la comparacion con respecto al operador que le entregamos por vectorizacion
#(de manera eficiente y en paralelo) a todos los valores en el array que especificamos
#¿Y ahora que hacemos una vez tenemos la mascara booleana? Se podria colocar encima de los datos
#y ocultar los que no sea desean. Esto se hace mediante la funcion where() ejemplo:
df.where(admit_mask).head()
```

```
##Como vemos la funcion nos mantiene todos los valores indexados, y aquello s que no cumplen se 
#presentan como NaN, estos datos no han sido eliminados de nuestro dataset
#Bueno si ahora no se quiere los datos NaN, usamos la función dropna() 
df.where(admit_mask).dropna().head()
```

```
##Como notamos ahora se ha saltado el index 5!
#Aunque es muy utíl, la verdad es que el where() no se usa muy seguido. En su lugar, los desarrolladores
#de Pandas han creado una sintaxis corta, que combina where() y dropnan, haciendo ambos a la vez
#sobrecargando el operador de indexacion. Ejemplo:

df[df['chance of admit'] > 0.7].head()
```

```
#Es un poco más dificl de leer, pero es muy común de leer en el codigo de otras personas.
#Haciendo uso del operador indexing podemos realizar dos tareas

#1. Puede ser llamado con un parametro string para llamar solo una columna
df['gre score'].head()
```

```
#O podemos enviarle una lista de columnas como cadenas:
df[['gre score','toefl score']].head()
```

```
# Y ahora podemos usarlo para enviar una mascara booleana
df[df['gre score']>320].head()
```

```
## Ahora pasemos al uso de multiples mascaras booleanas, es decir multiples criterios.
#En la mayoria de lenguajes de computación se realiza con un "and" si ambas condiciones deben ser verdaderas
#o con un "or" si cualquiera de las dos condiciones es aceptable.

#Desafortunadamente no se siente tan natural en Pandas, por ejemplo:
(df['chance of admit']>0.7) and (df['chance of admit']<0.9)
```

```
##Pero no funciona, esto es debido a que tenemos dos objetos en realidad, y Python no sabe como compararlos usango
#"AND" y/o "OR", en vez de ello Pandas ha sobreescrito el uso del simbolo "|" y del simbolo "&", para manejar esto
#por nosotros. ejemplo:
(df['chance of admit']>0.7)&(df['chance of admit']<0.9)
```

```
##Un error común entre nuevos ususarios de Pandas es el uso del operador ampersand "&" para comparaciones booleanas
#pero  no usar parentesis alrededor de los termino individales que nos interesen
df['chance of admit']>0.7&df['chance of admit']<0.9
```

```
##El problema anterior es que Python lo trata como un bit a bit con 0.7 y un dataFrame,
#cuando realmente lo que se quiere us una comparacion entre dataframes.

##Otra forma de solucionarlo es haciendo uso de funciones ya incorporadas, para evitar el uso de parentesis
df['chance of admit'].gt(0.7) & df['chance of admit'].lt(0.9)
```

```
##Estas funciones integradas directamente a objetos de Series y DataFrames, asi que
#se pueden encadenar sin necesidad de otros operadores, podemos elegir que es mejor para nosotros
df['chance of admit'].gt(0.7).lt(0.9)
```

# Indexing DataFrames

Como hemos visto, en Series y DataFrames se puede aplicar los indices, que basicamente son la eetiqueta de la fila en la que estamos, y en Pandas las filas corresponden al eje zero. Tambien tenemos la funcion set_index() para establecer el index deseado,esta funcion toma una lista de columnas y las promueve a indices. En esta seccion exploraremos mas de como funcionan la indexacion en pandas


```
## la funcion set_index() es un proceso destructivo, y no mantiene el indice actual. Si deseo mantener el indice
#actual se debe crear manualmente una nueva columna, y copiar en ella los valores del atributo index. Ejemplo:

import pandas as pd
df=pd.read_csv('/content/drive/MyDrive/Python_for_Data_Science/Week_2/datasets/Admission_Predict.csv',index_col=0)
df.head()
```

```
##Supongamos que no queremos indexar por Serial No, si no por 'Chance of Admit', asi que debemos
#presevar Serial No en una nueva columna. Esto se puede hacer usando el operador de indexacion 
#sobre el string de la etiqueta de la columna. EJEMPLO:

#1. Guardamos los indices en una nueva columna:
df['Serial Number']=df.index
#2. Asignamos el nuevo index:
df=df.set_index('Chance of Admit ')
df.head()
```

```
##Tambien podemos deshacernos del indice usando la funcion reset_index(). Esto hace que se promueva
#la columna indice a una columna normal y crea una columna indice nueva enumerada
df= df.reset_index()
df.head()
```

```
##Una de las caracteristicas de Pandas es la indexacion multinivel, de manera similar a los sistemas de 
#bases de datos relacionales. Para crear una indexacion multinivel, llamamos a set_index() y le damos
#una lista de columns que estamos interesados en promover a indice.

#Pandas buscará en orden a traves de esta lista, encontrando los distintos datos y los indices compuestos
#de esta forma. Un buen ejemplos de estos se suele encontrar en datos geograficos, en los que se ordenan 
#por region o demografía. Ejemplo con un archivo csv, de un censo en los Estados Unidos:

df= pd.read_csv('/content/drive/MyDrive/Python_for_Data_Science/Week_2/datasets/census.csv')
df.head()

```

```
## En esta base de datos hay dos niveles separados de resumen de informacion
#datos con informacion de todo el pais, y datos que contienen el resumen de cada estado
#Si se quierever una columna que tenga todos lo valores unicos. En este DataFrame, nosotros
#vemos los posibles valores para el SUMLEVEL usando el UNIQUE Function en el DataFrame. Esto
# es similar a SQL con el operador distinct.

#Entonces aquí podemos usar unique en sum level, para el DataFrame actual.

df['SUMLEV'].unique()
```

```
##Como podemos observar solo hay dos valores diferentes 40 y 50

##Ahora vamos a excluir todas las filas son resumenes al nivel de estado, y manetener
#Solo los datos del condado.

df=df[df['SUMLEV']==50] 
df.head()
```

```
##Aunque este dataset es intersante por diferentes razones
#Vamos reducir los datos solo a las estimaciones totales de poblacion y el numero
#total de nacimientos. Esto se puede realizar al crear una lista con los nombres de
#las columnas que queremos mantener, luego de esto proyectarlas y asiganarlas en un 
#DataFrame resultante en nuestra variable df. Ejemplo:

columns_to_keep= ['STNAME','CTYNAME','BIRTHS2010','BIRTHS2011','BIRTHS2012','BIRTHS2013',
                  'BIRTHS2014','BIRTHS2015','POPESTIMATE2010','POPESTIMATE2011',
                  'POPESTIMATE2012','POPESTIMATE2013','POPESTIMATE2014','POPESTIMATE2015']
df=df[columns_to_keep]                  
df.head()
```

```
##Los censos de USA desglosan las estimaciones por estado y condado, podemos cargar los datos
#y estableces el index para que sea la combinacion de los valores de estados y condado y ver 
#Pandas lo resuelve en un DataFrame. Se hace esto al crear una lista con los identificadores de
#columnas que se desean usar como indice y luego utilizar set_index() con esta lista, y asignando
#la salida segun corresponda.

df= df.set_index(['STNAME','CTYNAME'])
df.head()
```

```
##Una pregunta al ver esto es ¿Como realizamos consultas en este DataFrame? Nosotros
#vimos que un el atributo Loc puede tomar distintos argumentos. Y se puede consultar
#tanto en filas como en columnas. Cuando se usa MultiIndex, se debe proporcionar los 
#argumentos en orden desde el nivel que sea desea consultar. Dentro del index, cada 
#columna es denomida como un nivel y el nivel mas externo es la columna 0 (cero).

#Asi que si se deseara ver la poblacion del condado Washtenaw en el estado de Michigan,
#el primer argumento sería Michigan, y el segundo Washtenaw. Ejemplo:

df.loc['Michigan','Washtenaw County']
```

```
##Si lo que se desea es comparar dos condados por ejemplo Washtenaw and Wayne, podemos
#pasar una lista de tuplas describiendo los indices que se desean consultar con .loc.
#Debido a que se tiene MultiIndex de dos valores, el estado y el condado, debemos entregar 
#los dos valores como cada uno de los elementos de nuestra lista de filtrado. Cada tupla 
#debe tener dos elementos, el primer elemento es el primer index y de manera similar el segundo elemento 

#Ejemplo

df.loc[[('Michigan','Washtenaw County'),('Michigan','Wayne County')]]
```

Bueno, asi es como funcionan los indices jearquicos en pocas palabras, son una parte especial de la libreria de Pandas, que nos puede facilitar el razonamiento y manejo sobre los datos. No olvidar que los indices jerarquicos no son exclusivos de las filas, si transponemos las columnas podemos obetener indices jerarquicos en las columnas. Aunque en la realidad puede que no se tienda muchos a usar estos indices jerarquicos. Pero es una forma util y sofisticada de ver los datos sobre todo en forma tabular

# Missing Values (Valores faltantes)

Hemos visto como Pandas maneja los valores faltantes, usando None Type y Numpy NaN, los valores faltantes son muy comunes en la limpieza de datos. Y los valores faltantes pueden suceder por muchas razones, vamos a ver algunas de ellas acá.

Por ejemplo si en una encuesta alguien no respondió una pregunta, el valor faltante en realidad es una omisión. Esta clase de datos faltantes en realidad son llamados **Missing at Random** (perdidos al azar). Si hay otras variables que deberían ser usadas para predecir la variable que falta. En muchas ocasiones re realizan encuestas, y frencuentemente se encuentran valores faltantes, y por ejemplo a veces hablan del interes en el estudio realizado, a menudo tienen relación con otros datos como el genero o la etnia. Si no hay relacion con otras variables, entonces se llaman **Missing completely at Random(MCAR)**(datos faltantes completamente al azar).

Aquí dos ejemplos de los muchos que puede haber. Por ejemplo, los datos pueden faltar porque no fueron recolectados, ya sea por el responsable de recopilarlos o porque no tenía sentido que fueran recopilados. Otros ejemplo extremadamente común, cuando se empieza a unir DataFrames, a partir de diferentes fuentes, como unir una lista de personas en una universidad con una lista de oficinas en la universidad(Los estudiantes generalmente no tienen oficina).

Veamos algunas maneras de manejar estos casos en pandas.

# Manipulating DataFrame

En esta seccion se va realizar una limpieza tipica de datos, junto a la presentación de otras funciones de pandas:

```
import pandas as pd
##Ahora se carga el dataset que es una lista de presidentes de USA, obtenida de Wikipedia
df=pd.read_csv('/content/drive/MyDrive/Python_for_Data_Science/Week_2/datasets/presidents.csv')
df.head(10)
```

```
##Como vemos en el df, hay varias notas en la Born, las cuales podrían ocasionar problemas. Comencemos
#limpiando el nombre haciendo un first name y last name, mediante el uso de expresiones regulares Regex

df['First']=df['President']
#Ahora usamos la funcion 'replace()' para coventir el lastname(apellido) en una string vacio
df['First']=df['First'].replace('[ ].*','',regex=True)
df.head()
```

```
## Esto funciona pero es algo lento y burdo. El hecho de tener que generar una copia de la columna y luego
#arreglarla, hay otras manera de lidiar con ello en manera mas general. Una de ella es la funcion apply()
#Asi que eliminemos lo que acabamos de hacer.

del(df['First'])

##La funcion apply() va tomar algun parametro arbitratio que hayamos escrito y aplicarlo sobre una columna
#o sobre un DataFrame a traves de todas las filas y columnas. Se procede a escribir una función que nos 
#permita transformar en dos una columna

def splitname(row):
  #La fila es una objeto Serie la cual es indexada por el valor de una columna
  #Procedamos a extraer el nombre(firstname) en una nueva entrada de Series
  row['First']=row['President'].split(" ")[0]
  #Y lo mismo para el apellido
  row['Last']=row['President'].split(" ")[-1]
  return row
#Ahora hacemos uso de apply()
df=df.apply(splitname,axis='columns')
df.head()
```

```
##Si es menos burdo es cuestionable, pero Pandas posee un par de funciones extra
#bastante utiles, que serán tratados en otro apartado llamado extract().
del(df['First'])
del(df['Last'])

## extract() toma una expresion regular Regex como entrada y requiere que establezca
#especificamente los grupos de captura que correspondan a las columnas de salidas a 
# las cuales estamos interesados.

pattern="(^[\w]*)(?:.* )([\w]*$)"
#tres parentesis para establecer tres grupos, el primero al inicio, cualquier palabra, el segundo con signo "?"
#aclarando que no quiero que lo retorne, y quiero cualquier caracter seguido de un espacio. Y el tercero
#quiero cualquier nuemero de caracteres seguidos de un espacio en blanco pero anclado al final de la cadena

#ahora la funcion extract() esta dentro del atributo str del objeto Series, asi que se puede llamar usando
# Series.str.extract(pattern). Ejemplo:

df['President'].str.extract(pattern).head()
```

```
##Se ve bien pero si se nombran los grupos, se obtendra el nombre de las columnas. Actualicemos el pattern

pattern="(?P<First>^[\w]*)(?:.* )(?P<Last>[\w]*$)"
names=df['President'].str.extract(pattern).head()
names
```

```
##Ahora podriamos simplemente copiar esto dentro del DataFrame si lo queremos asi.
df['First']=names['First']
df['Last']=names['Last']
df.head()
```

Por tanto, se intuye que vale la pena mirar el modulo str de Pandas para otras funcionesque han sido escritos especificamente para limpiar cadenas y marcos de datos.
La documentacion puede ser encontrada en:
https://pandas.pydata.org/pandas-docs/stable/user_guide/text.html

```
##Ahora vamos a limpiar la columna de de nacimiento
#nos deshacemos de cualquier cosa que no pertenezca a la fehca
df["Born"]=df["Born"].str.extract("([\w]{3} [\w]{1,2}, [\w]{4})")
df['Born'].head()
```

```
##Esto ha limpiado nuestros datos, pero vayamos más allá, el tipo de esta columna
#es un objeto, y sabemos que eso es lo que Pandas usa cuando se trata de cadenas(strings).
#Pero Pandas en realidad tiene herramientas interesantes de fecha/hora, de hecho es una
#de las razones por la cual Wes McKinney pusos sus esfuerzos en la biblioteca fue para lidiar
#con transacciones financieras. Asi que si se estuviera hiendo mas alla con esto, se 
#actualizaria la biblioteca para escribir el tipo de dato tambien.  Para hacerlo se hace:

df['Born']=pd.to_datetime(df['Born'])
df['Born'].head()
```
